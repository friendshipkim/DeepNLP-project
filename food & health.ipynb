### v91 - 집에서 식사를 하는가

```python
food_health['v91'].value_counts()
```

* 보통 집에서 식사함 - 그럼 v90과 v91의 상관관계는?

```python
food_health.groupby(['v90','v91']).size().reset_index(name="count")
```

```python
v90_v91 = pd.crosstab(food_health['v90'],food_health['v91'])
v90_v91['Total'] = v90_v91.sum(axis=1)
v90_v91
```

```python
for i in range(1,6):
    v90_v91.loc[i] = v90_v91.loc[i].apply(lambda x: x/v90_v91.loc[i,'Total'])
```

```python
v90_v91
```

* 매일 아침 먹는 사람들은 약 66%가 집에서 매일 챙겨먹음 - 매우 부지런...
* 3~4번 먹는 사람들은 밖에서 사먹는 비율 비교적 높았음
